## Save PBMCs as mtx

```r
rna = seu@assays$RNA@counts
adt = seu@assays$CITE@counts
hto = seu@assays$HTO@counts

rna_adt_hto = rbind(rna, adt, hto)
```

```r
rna_adt_hto %>% dim
```

```r
rna_adt_hto %>% head
```

```r
## save to load as mtx
writeMM(rna_adt_hto, 
        '../../../Data/PerDataset/Sparks2023/PBMC_fromSeurat/matrix.mtx')


barcodes <- colnames(rna_adt_hto)
write_delim(as.data.frame(barcodes),
            '../../../Data/PerDataset/Sparks2023/PBMC_fromSeurat/barcodes.tsv',
           col_names = FALSE)

gene_names <- rownames(rna_adt_hto)
features <- data.frame("gene_id" = gene_names,"gene_name" = gene_names,type = "Gene Expression")
write_delim(as.data.frame(features),delim = "\t",
            '../../../Data/PerDataset/Sparks2023/PBMC_fromSeurat/features.tsv',
           col_names = FALSE)

# write_delim(filt@meta.data, '/data/srlab1/mcurtis/GSK/tcell_proliferation/TCellAnnotator/data/Sparks2023/T_fromSeurat_metadata.txt',
#             delim = '\t')
```

```r
write.table(x = seu@meta.data,quote = FALSE, row.names = TRUE,
            file =  '../../../Data/PerDataset/Sparks2023/PBMC_fromSeurat_metadata.txt',
           sep = '\t', )
```
