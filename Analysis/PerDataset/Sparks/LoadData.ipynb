```r
cluster_col = 'coarse.cell.type'
ctypes = c('CD4', 'CD8', 'gdT-Vd2', 'MAIT')

ind = rownames(seu@meta.data)[(seu@meta.data[[cluster_col]] %in% ctypes)]
```

```r
filt = subset(seu, cells = ind)
```

```r
filt %>% str(2)
```

```r
filt@assays$RNA@counts %>% head
```

```r
rna = filt@assays$RNA@counts
adt = filt@assays$CITE@counts
hto = filt@assays$HTO@counts

rna_adt_hto = rbind(rna, adt, hto)
```

```r
rna_adt_hto %>% dim
```

```r
rna_adt_hto %>% head
```

```r
## save to load as mtx
writeMM(rna_adt_hto, 
        '/data/srlab1/mcurtis/GSK/tcell_proliferation/TCellAnnotator/data/Sparks2023/T_fromSeurat_RNA_ADT_HTO/matrix.mtx')


barcodes <- colnames(rna_adt_hto)
write_delim(as.data.frame(barcodes),
            '/data/srlab1/mcurtis/GSK/tcell_proliferation/TCellAnnotator/data/Sparks2023/T_fromSeurat_RNA_ADT_HTO/barcodes.tsv',
           col_names = FALSE)

gene_names <- rownames(rna_adt_hto)
features <- data.frame("gene_id" = gene_names,"gene_name" = gene_names,type = "Gene Expression")
write_delim(as.data.frame(features),delim = "\t",
            '/data/srlab1/mcurtis/GSK/tcell_proliferation/TCellAnnotator/data/Sparks2023/T_fromSeurat_RNA_ADT_HTO/features.tsv',
           col_names = FALSE)

# write_delim(filt@meta.data, '/data/srlab1/mcurtis/GSK/tcell_proliferation/TCellAnnotator/data/Sparks2023/T_fromSeurat_metadata.txt',
#             delim = '\t')
```
